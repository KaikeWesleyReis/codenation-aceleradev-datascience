## Questão 1

Qual fração da variância consegue ser explicada pelo primeiro componente principal de `fifa`? Responda como um único float (entre 0 e 1) arredondado para três casas decimais.

```python
def q1():
    # Construir a PCA com 1 componente
    pca_obj = PCA(n_components=1).fit(fifa)
    # Retornar a variancia
    return float(pca_obj.explained_variance_ratio_[0].round(3))
```

```python
# Resposta
q1()
```

Quantos componentes principais precisamos para explicar 95% da variância total? Responda como un único escalar inteiro.

```python
def q2():
    # Construir a PCA buscando 95% de variancia explicada
    pca_obj = PCA(0.95).fit(fifa)
    # Retornar quantidade de componentes
    return int(pca_obj.n_components_)
```

## Questão 4

Realiza RFE com estimador de regressão linear para selecionar cinco variáveis, eliminando uma a uma. Quais são as variáveis selecionadas? Responda como uma lista de nomes de variáveis.

```python
def q4():
    # Definir x e y
    x = fifa.drop(columns="Overall")
    y = fifa["Overall"]
    # Construir modelo
    feat_selector = RFE(estimator=LinearRegression(), n_features_to_select=5).fit(x, y)
    # Pegar indices das variaveis selecionadas
    feat_selecionadas_idx = feat_selector.get_support(indices=True)
    # Retornar
    return list(x.columns[feat_selecionadas_idx])
```

```python
# Resposta
q4()
```
